### Multiple Linear Regression
Linear regression involving multiple variables is called “multiple linear regression” or multivariate linear regression.As per our dataset we can predict price based on surface and arrondissement of the house.So considering multiple variable will give some more good prediction as we dont get good accuracy with our simple linear regression where we consider to predict price of house baed on surface area. 

```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
#Import dataset
dataset = pd.read_csv('house_data.csv')
```

```python
#Explore Data
dataset.shape
```

```python
#Statistical details of data
dataset.describe()
```

```python
#Clean dataset_check which are the columns the contains NaN values in it :
dataset.isnull().any()
```

```python
#remove all the null values from the columns surface and arrondissement
dataset = dataset.fillna(method='ffill')
```

```python
#divide the data into “attributes” and “labels”. X variable contains all the attributes/features and y variable contains labels.
X = dataset[['surface', 'arrondissement']].values
y = dataset['price'].values
```

```python
# check the average value of the “price” column.
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(dataset['price'])
```

```python
#we split 80% of the data to training set while 20% of the data to test 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#train our model
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
#In multivariable regression,regression model has to find the most optimal coefficients for all the attributes
coeff_df = pd.DataFrame(regressor.coef_,columns=['Coefficient'])  
coeff_df

#0-surface 1-Arrondissement

#This means that for a area increase in “surface”, there is a increase of 36.66 units in the price of the house.
#Similarly, change in “Arrondissement“(From 1 to 10) results in an decrease of 45.31 units in the price of the house.
```

```python
#prediction on test data.
y_pred = regressor.predict(X_test)
```

```python
#difference between the actual value and predicted value
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
#plot the comparison of Actual and Predicted values
df1 = df.head(25)
df1.plot(kind='bar',figsize=(10,8))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
#Evaluate performance of algorithm as mentioned in prédiction_loyer file
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
